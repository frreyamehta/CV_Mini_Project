```python
import torch
import torch.nn as nn

# Example: Define your model class
class MyModel(nn.Module):
    def __init__(self):
        super(MyModel, self).__init__()
        # Example architecture
        self.conv1 = nn.Conv2d(3, 16, 3, 1)
        self.relu = nn.ReLU()
        self.fc = nn.Linear(16*30*30, 10)

    def forward(self, x):
        x = self.relu(self.conv1(x))
        x = x.view(x.size(0), -1)
        x = self.fc(x)
        return x

# Instantiate model
model = MyModel()

# Load pretrained weights
model.load_state_dict(torch.load("LGC_best.pth", map_location=torch.device('cpu')))

# Set to evaluation mode
model.eval()
```
